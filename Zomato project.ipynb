# Zomato Data Analysis Project

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
dataframe = pd.read_csv("Zomato data .csv")
dataframe
```

```python
dataframe
```

# Convert the data type of column - rate

```python
def handlerate(value):
    value = str(value).split('/')
    value = value[0]
    return float(value)
dataframe['rate']=dataframe['rate'].apply(handlerate)
dataframe.head()
```

```python
dataframe.info()
```

# Type of restorant

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'] )
plt.xlabel("type pd retorant")
```

# Conclusion - majority of the rastorants falls in dinning category

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result,c="green",marker="o")
plt.xlabel("type of restorant",size =20 , c= "red")
plt.ylabel("Votes",c ="red",size =20)
```

## Conclusion - Dining restorants recieved maximum votes

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'],bins = 5)
plt.title("ratings distribution")
plt.show()
```

# conclusion - the majority restorant received rating from 3.5 to 4

# Average order spending by couples

```python
dataframe.head()
```

```python
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

## conclusion - the majority of couples prefer restorants with an aproximate cost of 300 rupees. 

# which mode receives maximum ratings

```python
dataframe.head()
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x='online_order', y = 'rate',data=dataframe)
```

# conclusion - offline orders received lower rating in comparison to online order

```python
dataframe.head()
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap ="YlGnBu", fmt = 'd')
plt.title("Heatmap")
plt.xlabel("Online Orders")
plt.ylabel("Listed in (Types)")
plt.show()
```

# Conclusion - Dining restorants primarily accept offline ordeds,whereas cafes promarily receive online orders. This suggest that client prefers orders in person at restorants, but prefer online orders at cafes.
